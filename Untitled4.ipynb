```python
# create table jobs in SQLITE
import pandas as pd
import sqlite3
# connect to existing database
conn2 = sqlite3.connect('DWHospital.db')
cur = conn2.cursor()
```

```python
#create time dimension
cur.execute(''' CREATE TABLE time (
    date VARCHAR not null,
    month INTEGER not null,
    quarter INTEGER not null,
    year INTEGER not null,
    PRIMARY KEY (date)
)''')
```

```python
#create operational theatre table in database
cur.execute('''
CREATE TABLE operational_theatre (
    opth_id VARCHAR unique not null,
    ward_id VARCHAR not null,
    name VARCHAR unique not null,
    PRIMARY KEY (opth_id),
    FOREIGN KEY (ward_id) REFERENCES ward(ward_id)
)''')
```

```python
#create ward table in database
cur.execute('''
CREATE TABLE ward (
    ward_id VARCHAR unique not null,
    name VARCHAR unique not null,
    PRIMARY KEY (ward_id)
)''')
```

```python
#create operations fact table in database
cur.execute('''CREATE TABLE operation (
    operation_id VARCHAR unique not null,
    opth_id VARCHAR not null,
    admissions_id VARCHAR not null,
    charges INTEGER not null,
    type VARCHAR not null,
    PRIMARY KEY (operation_id),
    FOREIGN KEY (admissions_id) REFERENCES admissions(admissions_id),
    FOREIGN KEY (opth_id) REFERENCES operational_theatre(opth_id)
)''')
```

```python
#create admissions fact table in database
cur.execute('''
CREATE TABLE admissions (
    admissions_id INTEGER unique not null,
    ward_id VARCHAR not null,
    date VARCHAR not null,
    number_of_days INTEGER not null,
    cost INTEGER not null,
    PRIMARY KEY (admissions_id),
    FOREIGN KEY (date) REFERENCES time(date)
    FOREIGN KEY (ward_id) REFERENCES ward(ward_id)
    )
''')
```

```python
#create diagnosis table in database
cur.execute('''
CREATE TABLE diagnosis (
    diagnosis_id VARCHAR unique not null,
    type VARCHAR not null,
    name VARCHAR not null,
    PRIMARY KEY (diagnosis_id)
    )
''')
```

```python
#create the secondary diagnosis fact table
cur.execute('''
CREATE TABLE secondary_diagnosis_admissions (
    admissions_id INTEGER unique not null,
    diagnosis_id VARCHAR not null,
    number_of_days INTEGER not null,
    cost INTEGER not null,
    secondary_diagnosis_id VARCHAR unique not null,
    PRIMARY KEY (admissions_id),
    FOREIGN KEY (admissions_id) REFERENCES admissions(admissions_id),
    FOREIGN KEY (diagnosis_id) REFERENCES diagnosis(diagnosis_id)
    )
''')
```

```python
#checking the successful creation of all tables
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cur.fetchall()
print("Tables in DWHospital.db:")
for table in tables:
    print(table[0])
```

```python
#save what's been done
conn2.commit()
```

```python
#final materialized view for Q1
cur.execute('''
CREATE VIEW Q1 AS
SELECT
    w.name AS Ward,
    t.year AS Year,
    t.quarter AS Quarter,
    SUM(o.charges) AS Total_Charges,
    SUM(a.number_of_days) AS Total_LOS,
    SUM(a.cost) AS Total_Admissions_Cost
FROM admissions a, ward w, time t
LEFT JOIN operation o ON a.admissions_id = o.admissions_id
WHERE a.ward_id = w.ward_id
AND a.date = t.date
GROUP BY w.name, t.year, t.quarter
ORDER BY w.name ASC, t.year ASC, t.quarter ASC;
''')
```

```python
#show the results of the view for q1
cur.execute("SELECT * FROM Q1")
results = cur.fetchall()

# grab the column headers
column_names = [description[0] for description in cur.description]

#put in dataframe for readability
q1_df = pd.DataFrame(results, columns=column_names)
q1_df
```

```python
#Q2:  For each Ward  Year,quarter report the charges  per operation, length of stay and admission cost of Secondary Diagnosis.
cur.execute('''
CREATE VIEW v2_Q2 AS
SELECT
    w.name AS Ward,
    t.year AS Year,
    t.quarter AS Quarter,
    SUM(o.charges) AS Total_Charges,
    SUM(sda.number_of_days) AS Total_LOS,
    SUM(sda.cost) AS Secondary_Diagnosis_Admissions_Cost
FROM admissions a, ward w, time t, secondary_diagnosis_admissions sda
LEFT JOIN operation o ON a.admissions_id = o.admissions_id
WHERE a.ward_id = w.ward_id
AND a.date = t.date
AND sda.admissions_id = a.admissions_id
GROUP BY w.name, t.year, t.quarter
ORDER BY w.name ASC, t.year ASC, t.quarter ASC;
''')
```

```python
#show the results of the view for q2
cur.execute("SELECT * FROM v2_Q2")
results = cur.fetchall()

# grab the column headers
column_names = [description[0] for description in cur.description]

#put in dataframe for readability
q2_df = pd.DataFrame(results, columns=column_names)
q2_df
```
